```python
import pandas as pd

# Read the CSV files into DataFrames
df_athletes = pd.read_csv('/content/athlete_events[1].csv')
df_noc = pd.read_csv('/content/noc_regions[1].csv')

# Display the first few rows to check
print(df_athletes.head())
print(df_noc.head())
```

```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('/content/olympic.db')
cursor = conn.cursor()

# Create tables in SQLite
cursor.execute('''
CREATE TABLE IF NOT EXISTS athletes (
    id INTEGER PRIMARY KEY,
    name TEXT,
    sex TEXT,
    age INTEGER,
    height INTEGER,
    weight INTEGER,
    team TEXT,
    noc TEXT,
    games TEXT,
    year INTEGER,
    season TEXT,
    city TEXT,
    sport TEXT,
    event TEXT,
    medal TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS nocs (
    noc TEXT PRIMARY KEY,
    region TEXT,
    notes TEXT
)
''')

# Insert data into tables
df_athletes.to_sql('athletes', conn, if_exists='replace', index=False)
df_noc.to_sql('nocs', conn, if_exists='replace', index=False)

# Verify the data was inserted
print(pd.read_sql_query('SELECT * FROM athletes LIMIT 5', conn))
print(pd.read_sql_query('SELECT * FROM nocs LIMIT 5', conn))
```

```python
# Example query: List all athletes who have won a gold medal
query = '''
SELECT DISTINCT name
FROM athletes
WHERE medal = 'Gold'
'''

# Execute the query and fetch the results
results = pd.read_sql_query(query, conn)
print(results)

# Close the connection
conn.close()
```

```python
import sqlite3
import pandas as pd
from google.colab import files

# Upload the CSV files
uploaded = files.upload()

# Load the dataset into pandas DataFrames
df_athletes = pd.read_csv('/content/athlete_events[1].csv')
df_noc = pd.read_csv('/content/noc_regions[1].csv')

# Create a connection to the SQLite database
conn = sqlite3.connect('olympics.db')

# Create tables in SQLite
conn.execute('''
CREATE TABLE IF NOT EXISTS athletes (
    ID INTEGER,
    Name TEXT,
    Sex TEXT,
    Age INTEGER,
    Height INTEGER,
    Weight INTEGER,
    Team TEXT,
    NOC TEXT,
    Games TEXT,
    Year INTEGER,
    Season TEXT,
    City TEXT,
    Sport TEXT,
    Event TEXT,
    Medal TEXT
)
''')

conn.execute('''
CREATE TABLE IF NOT EXISTS nocs (
    NOC TEXT PRIMARY KEY,
    region TEXT,
    notes TEXT
)
''')

# Insert data into tables
df_athletes.to_sql('athletes', conn, if_exists='replace', index=False)
df_noc.to_sql('nocs', conn, if_exists='replace', index=False)

# Example query: List All Athletes Who Have Won More Than One Medal
query = '''
SELECT Team, COUNT(Medal) AS Gold_Medals
FROM athletes
WHERE Medal = 'Gold'
GROUP BY Team
ORDER BY Gold_Medals DESC
LIMIT 10;
'''


# Execute the query and fetch the results
results = pd.read_sql_query(query, conn)
print(results)

# Close the connection
conn.close()
```
